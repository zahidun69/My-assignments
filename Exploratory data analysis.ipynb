```python
'''Exploratory Data Analysis on a Dataset'''
#importing data file
import pandas as pd
df=pd.read_csv("Cardiotocographic.csv")
df
```

```python
#Handling missing values
#STEP-1:Identifying the Missing values
df.isnull().sum()
```

```python
#Imputation of the misssing values using mean.
df.fillna(df.mean(), inplace=True)
df
```

```python
#Handling outliers using z-score calculations 
# Calculate the z-score for each column
z_scores = (df - df.mean()) / df.std()

# Define the threshold for outliers
threshold = 3

# Select rows where the absolute z-score is less than the threshold
df = df[(abs(z_scores) < threshold).all(axis=1)]

# Print the cleaned DataFrame
df
```

```python
#2.Statistical Summary:
df.describe()
```

```python
#3.Data Visualization:
#creating histogram for all numerical columns to understand the distribution of data.
import matplotlib.pyplot as plt

# Create histograms for each numerical column
for col in df.select_dtypes(include=['number']):
    plt.hist(df[col], bins=10)
    plt.title(f"Histogram of {col}")
    plt.show()
```

```python
# Separating the numerical and the categorical columns.
numerical_cols = df.select_dtypes(include=["number"])
categorical_cols = df.select_dtypes(exclude=["number"])

# Generating scatter plots to understand the pairs of numerical variables.
import matplotlib.pyplot as plt
for i, col1 in enumerate(numerical_cols):
    for j, col2 in enumerate(numerical_cols):
        if i > j:
            plt.scatter(df[col1], df[col2])
            plt.xlabel(col1)
            plt.ylabel(col2)
            plt.show()
```

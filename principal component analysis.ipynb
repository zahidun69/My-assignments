```python
'''  PCA   '''
#UPLOADING FILE
import pandas as pd
df=pd.read_csv("wine.csv")
df.head()
```

```python
#Correlation between variables
print(df.describe())
corr_matrix=pd.DataFrame(df.corr())
corr_matrix
```

```python
#finding high correlation variables
for col in corr_matrix.columns:
  highest_correlation = corr_matrix[col].sort_values(ascending=False).index[1]
  print(f"{col}: {highest_correlation}")
```

```python
#histogram for all numerical variables
import matplotlib.pyplot as plt

for col in df.select_dtypes(include=['int64', 'float64']):
  plt.hist(df[col])
  plt.title(col)
  plt.show()
```

```python
#standardization of continuous variables
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
df_scaled = pd.DataFrame(df_scaled, columns=df.columns)
df_scaled.head()
```

```python
#Implementing PCA 
from sklearn.decomposition import PCA
pca = PCA()
pc = pca.fit_transform(df_scaled)
pc = pd.DataFrame(pc)
pc.head()
```

```python
#checking variance ratio on pca
for i in range(0,14):
  print(pca.explained_variance_ratio_[i])
```

```python
sum(pca.explained_variance_ratio_*100)
```

```python
X_new=pc.iloc[:,0:2]
X_new
```

```python
#k-mean clustering for original dataset
X=df.iloc[:,0:]
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
scores = []

for i in range(2,10):
    cluster = KMeans(n_clusters=i,n_init=20)
    df["kmeans"] = cluster.fit_predict(X)
    Y = df["kmeans"]
    scores.append(silhouette_score(X,Y))
    
    
import matplotlib.pyplot as plt
plt.scatter(range(2,10),scores)
plt.plot(range(2,10),scores,color='red')
plt.xlabel("cluster numbers")   
plt.ylabel("silhoutte scores")
plt.show()
```

```python
#experiment k-mean cluster with different parametes
from sklearn.metrics import silhouette_score

scores = []

for i in range(2,10):
    cluster = KMeans(n_clusters=i,n_init=20)
    df["kmeans"] = cluster.fit_predict(X_new)
    Y = df["kmeans"]
    scores.append(silhouette_score(X_new,Y))
    
    
import matplotlib.pyplot as plt
plt.scatter(range(2,10),scores)
plt.plot(range(2,10),scores,color='red')
plt.xlabel("cluster numbers")   
plt.ylabel("silhoutte scores")
plt.show()
```

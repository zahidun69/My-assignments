```python
#apply standardization for numerical variables

from sklearn.preprocessing import StandardScaler

# Selecting numerical columns
numerical_cols = df.select_dtypes(include=['float', 'int']).columns

# Creating a StandardScaler object
scaler = StandardScaler()

# Fitting and transforming the numerical columns
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
df[numerical_cols]
```

```python
#Apply One-Hot Encoding to categorical variables with less than 5 categories.

import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Selecting categorical columns with less than 5 categories
categorical_cols = [col for col in df.select_dtypes(include=['object']).columns if df[col].nunique() < 5]

# Creating a OneHotEncoder object
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')

# Fitting and transforming the categorical columns
encoded_data = encoder.fit_transform(df[categorical_cols])

# Creating a DataFrame from the encoded data
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(categorical_cols))

# Concatenating the encoded DataFrame with the original DataFrame
df = pd.concat([df.drop(categorical_cols, axis=1), encoded_df], axis=1)
df
```

```python
#Use Label Encoding for categorical variables with more than 5 categories.

from sklearn.preprocessing import LabelEncoder

# Selecting categorical columns with more than 5 categories
categorical_cols = [col for col in df.select_dtypes(include=['object']).columns if df[col].nunique() >= 5]

# Creating a LabelEncoder object
label_encoder = LabelEncoder()

# Fitting and transforming the categorical columns
for col in categorical_cols:
    df[col] = label_encoder.fit_transform(df[col])
df
```

```python
#Create at least 2 new features that could be beneficial for the model. Explain the rationale behind your choices.

# Creating a new feature 'hours_per_week_squared'
df['hours_per_week_squared'] = df['hours_per_week'] ** 2
# Rationale: This feature captures potential non-linear relationships between 
# working hours and income. It could be that people working extremely long hours
# or very few hours have different income patterns than those working a standard week.
 
# Creating a new feature 'is_high_education'
df['is_high_education'] = df['education'].apply(lambda x: 1 if x in [12, 13, 14, 15] else 0) 
# Rationale: Grouping higher education levels together might provide a stronger 
# signal for income prediction compared to treating each level individually. 
# This assumes a general trend of higher income with higher education.
df
```

```python
#Apply a transformation (e.g., log transformation) to at least one skewed numerical feature and justify your choice.

import numpy as np

# Checking the distribution of 'capital-gain'
print(df['capital_gain'].skew())  # High positive skewness

# Applying log transformation to 'capital-gain'
df['capital-gain_log'] = np.log1p(df['capital_gain'])

# Checking the distribution after transformation
print(df['capital-gain_log'].skew())  # Reduced skewness

# Justification:
# The 'capital-gain' feature is highly right-skewed, with a long tail of high values. 
# This can negatively impact some machine learning models that assume a normal distribution. 
# Applying a log transformation helps reduce the skewness, making the distribution more symmetrical 
# and potentially improving model performance. 
```

```python
#Use the Isolation Forest algorithm to identify and remove outliers:

from sklearn.ensemble import IsolationForest

# Creating an IsolationForest model
iso_forest = IsolationForest(contamination=0.01)  # Adjust contamination as needed

# Fitting the model and predicting outliers
outliers = iso_forest.fit_predict(df)

# Removing outliers from the DataFrame
df = df[outliers != -1]
df
```

```python
Discuss how outliers can affect model performance (e.g., biasing the model or reducing generalization) in df dataset from above isolation forest

# Outliers can significantly impact model performance in several ways:

# 1. Bias: 
# Outliers, being extreme values, can disproportionately influence the model's parameters during training. 
# This can lead to a biased model that performs well on the outliers but poorly on the majority of the data. 
# For instance, in a linear regr ession model, a single outlier can significantly alter the slope of the regression line.

# 2. Reduced Generalization:
# Outliers can cause the model to overfit to the training data, reducing its ability to generalize to unseen data. 
# The model might learn to capture the specific patterns of the outliers, which might not be representative of the overall population.

# 3. Increased Variance:
# In some models, outliers can increase the variance of the model's predictions, making them less reliable. 
# This is particularly true for models sensitive to extreme values, such as those based on distance calculations (e.g., k-nearest neighbors).

# In the context of the provided Isolation Forest model, removing the outliers (identified as -1) aims to mitigate these issues. 
# By focusing on the "normal" data points, the model is expected to learn more representative patterns and improve its generalization performance on unseen data.

# However, it's crucial to note that outlier removal is not always the best solution. 
# In some cases, outliers might contain valuable information or indicate rare but important events. 
# The decision to remove outliers should be based on a careful analysis of the data and the specific problem being addressed.
```

```python


import ppscore as pps

# Calculate the Predictive Power Score matrix
matrix = pps.matrix(df)

# Display the matrix
print(matrix)

# Calculate the correlation matrix
corr_matrix = df.corr()

# Display the correlation matrix
print(corr_matrix)

# Compare PPS and correlation for selected features
print("PPS for 'education' predicting 'income':", matrix.loc['education', 'occupation'])
print("Correlation between 'education' and 'income':", corr_matrix.loc['education', 'occupation'])
```

```python
'' Discussion:

 The Predictive Power Score (PPS) and correlation matrix provide different perspectives on feature relationships.

- Correlation: Measures the linear association between two variables. A high correlation indicates a strong linear relationship, 
   but it doesn't necessarily imply that one variable predicts the other well.

- PPS: Measures how well one feature can predict another, considering both linear and non-linear relationships. 
  A high PPS indicates strong predictive power, even if the correlation is low.

 By comparing PPS and correlation, we gain a more comprehensive understanding of feature relationships and their potential for predictive modeling. 
 PPS can reveal valuable non-linear relationships that correlation might overlook, guiding feature selection and model development. 
```
